```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
ax=sns.violinplot(x="total_bill",data=tips)
plt.show()
```

```python
fig, ax = plt.subplots()
tips = pd.read_csv("tips.csv")
ax.violinplot(tips["total_bill"], vert=False)

plt.show()
```

```python
tips = sns.load_dataset("tips")
ax = sns.violinplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
ax=sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips)
```

```python
ax=sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips, split=True)
```

```python
ax=sns.violinplot(x="day", y="total_bill", hue="smoker", split=True, data=tips, inner="quartile")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
ax = sns.catplot(x="time", y="total_bill", hue="smoker", data=tips)
plt.show()
```

```python
ax=sns.catplot(x="time", y="total_bill", hue="smoker", col="sex", data=tips)
```

```python
ax=sns.catplot(x="time", y="total_bill", hue="smoker", kind="violin", data=tips)
plt.show()
```

```python
ax=sns.catplot(x="time", y="total_bill", hue="smoker", kind="violin", data=tips)
```

```python
ax=sns.countplot(x="time", hue="smoker", data=tips)
plt.show()
```

```python
ax=sns.countplot(y="time", hue="smoker", data=tips)
```

```python
ax=sns.swarmplot(x=tips["total_bill"])
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.lineplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.show()
```

```python
sns.scatterplot(x="time", y="tip",hue="sex", data=tips)
plt.show()
```

```python
sns.scatterplot(x="time",y="tip",hue="sex",style="smoker", data=tips)
plt.show()
```

```python
sns.kdeplot(tips.total_bill,tips.tip,cmap="Reds")
plt.show()
```

```python
corr=tips.corr()
sns.heatmap(corr)
plt.show()
```
